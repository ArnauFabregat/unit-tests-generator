```python
import json
import sys
sys.path.insert(0, "..")

from src.raggraph.walker import build_graph_from_directory
```

```python
g = build_graph_from_directory(code_path="../src", save_graph=True, save_path="../data/repo_graph.graphml")
```

```python
print("Nodes:", g.number_of_nodes())
print("Edges:", g.number_of_edges())
```

```python
# Print nodes for inspection
for n, data in list(g.nodes(data=True)):
    print(n, "\n\t", json.dumps(data, indent=2))
    print()
```

```python
# Print edges for inspection
for src, dest, data in list(g.edges(data=True)):
    print(src, "--", data["rel"], "-->", dest)
    print()
```
